```python
#unpivoting the data and creating a column called Date and saving it in a new variable

confirmed_data1=pd.melt(confirmed_data, id_vars=["Province/State","Country/Region", "Lat","Long"],var_name=["Date"])
deaths_data1=pd.melt(deaths_data, id_vars=["Province/State","Country/Region", "Lat","Long"],var_name=["Date"])
recovered_data1=pd.melt(recovered_data, id_vars=["Province/State","Country/Region", "Lat","Long"],var_name=["Date"])
```

```python
# checking for null values 

confirmed_data1.isnull().sum()
```

```python
# Replacing the null values in Provice/State with values from Country/Region

confirmed_data1['Province/State'].fillna(confirmed_data1['Country/Region'],inplace=True)
deaths_data1['Province/State'].fillna(deaths_data1['Country/Region'],inplace=True)
recovered_data1['Province/State'].fillna(recovered_data1['Country/Region'],inplace=True)
```

```python
confirmed_data1.isnull().sum()
```

```python
# Converting Date colume to Datetime

confirmed_data1['Date']=pd.to_datetime(confirmed_data1['Date'])
deaths_data1['Date']=pd.to_datetime(deaths_data1['Date'])
recovered_data1['Date']=pd.to_datetime(recovered_data1['Date'])
```

```python
# But i will be using this method for changing the value column to Confirmed, Deaths and Recovered both 
#note both methods are allowed in mpython 

confirmed_data1.rename(columns={'value':'Confirmed'}, inplace=True)
deaths_data1.rename(columns={'value':'Deaths'}, inplace=True)
recovered_data1.rename(columns={'value':'Recovered'}, inplace=True)
deaths_data1.head()
```

```python
#joining three dataset into one single data set 
# Usimg the merge  function firstly i joined the confirmed_data1 with the deaths_data1 
#i choosed the columns i want in the second data and conduct a outer join 
#left on his the first data and right on is the second data 
#full join becuse even if the country donot have cases on a day the till have thier roll

join=confirmed_data1.merge(deaths_data1[["Province/State","Country/Region", "Date","Deaths"]],
                           how="left",
                           left_on=["Province/State","Country/Region", "Date"],right_on=["Province/State","Country/Region", "Date"])


fulljoin=join.merge(recovered_data1[["Province/State","Country/Region", "Date","Recovered"]],
                    how="left",
                    left_on=["Province/State","Country/Region", "Date"],right_on=["Province/State","Country/Region", "Date"])


fulljoin.head()
```

```python
fulljoin.shape
```

```python
fulljoin.isnull().sum()
```

```python
# Copying the fulljoin and save in a new variable = newfulljoin

newfulljoin=fulljoin.copy()

# Getting the previous day using the Timedelta, renaming the Confirmed, Deaths and Recovered. Merging fulljoin and newfulljoin.

newfulljoin['Date - 1']=newfulljoin['Date']+pd.Timedelta(days=1)
newfulljoin.rename(columns={'Confirmed':'Confirmed -1','Deaths':'Deaths -1','Recovered':'Recovered -1','Date':'Date minus 1'},inplace=True)

newfulljoin2=fulljoin.merge(newfulljoin[['Province/State','Country/Region','Confirmed -1','Deaths -1','Recovered -1','Date - 1','Date minus 1']],how='left',left_on=['Province/State','Country/Region', 'Date'],right_on=['Province/State','Country/Region', 'Date - 1'])


newfulljoin2.head()
```

```python
# Creating Three new colums i.e. Confirmed Daily,Deaths Daily and Recovered Daily


newfulljoin2['Confirmed Daily']=newfulljoin2['Confirmed']-newfulljoin2['Confirmed -1']
newfulljoin2['Deaths Daily']=newfulljoin2['Deaths']-newfulljoin2['Deaths -1']
newfulljoin2['Recovered Daily']=newfulljoin2['Recovered']-newfulljoin2['Recovered -1']

newfulljoin2.head()
```

```python
# Filling the 2020-01-22 Confirmed Daily column with value from Confirmed 2020-01-22 inorder to remove the null values.

newfulljoin2['Confirmed Daily'].loc[newfulljoin2['Date']=='2020-01-22'] = newfulljoin2['Confirmed']
newfulljoin2['Deaths Daily'].loc[newfulljoin2['Date']=='2020-01-22']= newfulljoin2['Deaths']
newfulljoin2['Recovered Daily'].loc[newfulljoin2['Date']=='2020-01-22']= newfulljoin2['Recovered']

newfulljoin2.head()
```
